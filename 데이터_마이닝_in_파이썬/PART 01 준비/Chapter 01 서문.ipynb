# Part01 준비

## Chapter 01 서문

### 1.1 비즈니스 애널리틱스 정의

- 비즈니스 애널리틱스(BA) = 데이터 분석
  - 의사결정에 필요한 정량적인 데이터를 산출하는 업무이자 기술
  - (복잡한) 데이터 분석 방법
      - 데이터 탐색, 변수 관계 측정 및 설명, 변수 값 예측
  - '과거 상황과 현재 상황'을 이해하기 위한 데이터 시각화 및 보고
  - 과거에 주로 정적 보고서를 작성하던 사용자가 실시간 데이터에 접근할 수 있을 뿐만 아니라, 직접 상호 작용할 수 있는 인터랙티브 대시보드 제작하는 등의 사용자 친화적 및 효과적인 툴 사용
  - 대규모의 복잡한 데이터베이스에 숨겨진 정보 빠르게 찾기 가능
  - ex) 회귀모델 : '평균적인' 관계(공고와 매출액)를 기술하거나 정량화하고, 새로운 사건(신규 환자가 치료약에 양성 반응할지 여부)을 예측할 수 있으며, 미래 변수 값(다음 주의 웹 트래픽양)을 에상 가능

### 1.2 데이터 마이닝 정의

- 데이터 마이닝
  - 데이터 시각화 +< 데이터 분석
    - ex) 의사결정을 자동화하는 방식으로 정보를 제공하는 통계 및 머신러닝 방법
    - ex) 고객을 차별화된 마케팅 판촉을 받는 각기 다른 '성향'으로 군집화한 후, 새로운 잠재 고객을 이 군집 중 하나로 할당
  - 강력하고 자동화된 기능으로 대용량 데이터를 처리 및 가치 추출

### 1.3 데이터 마이닝 관련 용어

- 예측 분석, 예측 모델링
- 머신러닝 : 개개인의 값을 예측 (=/ 추론)
  - 계층적 또는 반복적으로 직접 데이터로부터 학습하는 알고리즘
    - 선형 회귀 모델, k-초근접 이웃 알고리즘 : 주어진 레코드는 모든 레코드에 적용되는 전체 선형 방정식에 따라 선형 회귀에 의해 처리됨 (기록은 소수의 근접 기록 값에 따라 분류됨)
- 데이터 마이닝 : 통계학과 머신러닝(인공지능)의 결합
  - 단점 : 과적합 위험(모델이 신호뿐만 아니라 노이즈까지도 분석)


### 1.4 빅데이터

- 빅데이터
  - 데이터 마이닝과 함께 성장
  - 특징(4V)
    - 볼륨(volume) : 데이터양
    - 속도(velocity) : 유동률(데이터가 생성되거나 변경되는 속도)
    - 다양성(variety) : 생성되는 다양한 유형의 데이터(화폐, 날짜, 숫자, 텍스트)
    - 정확성(veracity) : 데이터가 유기적인 분산 프로세스에 의해 생성되며, 어떤 연구를 위해 수집된 데이터에 적용되는 통제 또는 품질 검사의 대상이 아님

### 1.5 데이터 사이언스

- 데이터 사이언스 : 통계, 머신러닝, 수학, 프로그래밍, 비즈니스, IT 분야 기술 혼합한 것
- 목표 : 배포된 시스템에 최종적으로 연결되는 통계 및 머신러닝 모델을 개발하는 것
  - 어떤 종류의 데이터 및 문제를 풀기 위해 어떤 방법 사용?
  - 분석 방법들은 어떻게 작동?
  - 각 분석 방법의 요구 사항, 장점, 약점?
  - 각 분석 방법의 성과 테스트는 어떻게?


### 1.6 다양한 분석 방법이 존재하는 이유

- 다양한 예측 및 분류 방법
- 방법의 유용성에 영향을 끼치는 요소
  - 데이터셋의 크기
  - 데이터에 존재하는 패턴의 유형
  - 데이터가 해당 방법의 일부 기본 가정을 충족하는지 여부
  - 데이터의 노이즈 양
  - 분석의 특정 목표

### 1.7 용어와 표기법

- 예측하는 변수 : 결과 변수=목표 변수=종속 변수=응답
- P(A|B) : 사건 B가 발생했을 때, 사건 A가 발생할 조건부 확률
- 검증 데이터 : 모델이 얼마나 적합한지 테스트 후, 모델을 조정해 보면서 최적 모델을 결정하는 데 사용한 일부 데이터
- 결과 변수 : '반응' 참조
- 관측
  - 측정된 것들(고객, 거래 등)의 분석 단위, 인스턴스, 표본, 예제, 케이스, 레코드, 패턴 또는 행
  - 스프레드시트에서 각 행은 하나를 나타내고, 각 열은 변수를 나타냄
  - 표본
- 레코드 : '관측' 참조
- 모델 : 데이터셋에 적용되는 알고리즘(사용자가 조정 가능한 모수들을 가짐)
- 목표 변수 : '반응' 참조
- 반응
  - Y
  - 지도 학습에서 예측되는 변수, 종속 변수, 출력 변수, 목표 변수, 결과 변수
- 범주형 변수 : 여러 고정값 중 하나를 취하는 변수 (ex. 정시, 연착, 취소)
- 변수 : 기록들을 측정하는 척도, 입력 변수(X) 및 출력 변수(Y) 포함
- 비지도 학습 : 분석 대상의 결과값을 예측하는 것 외에 데이터에 대해 뭔가를 알아보려는 분석
- 성공 클래스 : 이진법 결과 내의 관심 클래스(결과 변수 구매/비구매의 구매자)
- 속성 : '예측 변수' 참조
- 신뢰 : 반대의 표본을 선택했을 때의 추정 오차
- 신뢰도 : A와 B를 구매했다면 C도 구매한 것이라는 조건부 확률
- 알고리즘 : 분류 트리, 판별 분석 등 특정 데이터 마이닝 기술을 실행하기 위한 세부 과정
- 예측 : 연속적인 출력 변수의 예측값, 추정
- 예측 변수 :
  - X
  - 예측 모델에서 변수
  - 특성, 입력 변수, 독립 변수, 데이터베이스 측면의 필드
- 요인 변수 : '범주형 변수' 참조
- 입력 변수 : '예측 변수' 참조
- 점수 : 예측된 값이나 계층, 새로운 데이터를 점수화한다는 것은 학습(훈련) 데이터로 개발된 모델을 사용해 새로운 데이터의 결과 값을 예측한다는 의미
- 종속 변수 : '응답' 참조
- 지도 학습 : 결과 변수를 아는 레코드를 알고리즘(로지스틱 회귀, 회귀 트리 등)에 제공하는 과정, 알고리즘은 결과 변수를 모르는 상태에서 새로운 레코드들을 이용해 이 값으 예측하는 방법을 '배움'
- 추정 : '예측' 참조
- 출력 변수 : '반응' 참조
- 케이스 '관측' 참조
- 테스트 데이터 : 모델 구축과 선택 과정이라는 마지막 단계에서 최종 모델이 새로 추가되는 데이터를 얼마나 잘 예측하는지 테스트에 사용
- 특성 : '예측 변수' 참조
- 표본 통계 : 관측들의 집합, 한 가지 관측
- 프로파일 : 관측 결과 측정된 값들(사람의 키 몸무게, 나이 등)
- 학습 데이터 : 모델링하는 데 사용되는 일부 데이터
- 홀드아웃 데이터(표본)
  - 모델의 수행을 테스트할 때 쓰임
  - 검증 데이터셋
  - cf. 홀드아웃 표본 = 테스트 데이터셋

### 1.8 로드맵

- 1부(1~2장) : 데이터 마이닝의 총론과 구성 요소
- 2부(3~4장) : 데이터 탐색과 차원 축소 (초기 단계)
- 3부(5장)
  - 성능 평가 설명
  - 예측 수행 척도부터 오분류 비용까지
  - 추후 평가와 지도 학습 방법에 중요
- 4부(6~13장)
  - 인기 있는 지도 학습 방법들(분류, 예측)
    - 알고리즘의 난이도, 인기도, 이해도에 따라 구성
    - 마지막 장 : 앙상블 및 여러 기법의 조합
- 5부(14~15장)
  - 관계를 비지도 학습 방법으로 분석
  - 연관성 귝칙과 협업 필터링 설명
  - 군집 분석 설명
- 6부(16~18장)
  - 시계열 예측
    - 시계열의 취급과 이해
    - 시계열에서 많이 쓰이는 예측 기법(회귀 기반 예측, 평활법)
  - 7부(19~20장)
  - 데이터 애널리틱스 주제(특이한 구조 : 소셜 네트워크 분석, 텍스트 마이닝)
    - 특이한 데이터 구조에 데이터 마이닝을 적용한 구조
- 8부(21장) : 관련 사례들
- 순서
  - 1~3부(1~5장)
  - 4~5부(6~15장)
  - 6부의 16장
  - 6부의 17~18장
  - 7부~8부(19~21장)
- 파이썬
  - 데이터 분석, 데이터 시각화, 머신러닝을 위한 다양한 패키지 사용 가능
  - 분류, 예측, 텍스트 마이닝, 예상, 데이터 탐색 및 축소
  - 예측 모델링
    - 신경망, 분류 및 회귀 트리, k-최근접 이웃 분류, 나이브 베이즈 분류, 로지스틱 회귀, 선형 회귀 및 판별 분석과 같은 다양한 지도 학습 데이터 마이닝 툴 제공
  - 비지도 학습 알고리즘
    - 연관 규칙, 협업 필터링, 주성분 분석, K-평균 군집화, 계층적 군집화, 시각화 툴 및 데이터 처리 유틸리티
  - 동일한 방법이 다양한 패키지로 구현되는 경우 다수
